```python
import zipfile
import os
```

```python
files = ['../data/images/' + file for file in os.listdir('../data/images') if '.zip' in file]
```

```python
for file in files:
    out_dir = file.replace('.zip', '')
    with zipfile.ZipFile(file, 'r') as zip_ref:
        zip_ref.extractall(out_dir)
    os.remove(file)
```
