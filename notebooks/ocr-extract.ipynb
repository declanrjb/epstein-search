```python
from PIL import Image
import pytesseract as pt
import os
import json
```

```python
def extract_data(file):
    img = Image.open(file)
    raw_text = pt.image_to_string(img)
    return {
            'file': file,
            'text': raw_text,
            'sample': raw_text[0:200],
            'file_stem': file.split('/')[-1]
        }
```

```python
folders = ['../data/images/' + folder + '/IMAGES' for folder in os.listdir('../data/images') if folder != '.DS_Store']

for i in range(0, len(folders)):
    folder = folders[i]
    subfolders = [f'{folder}/{sub}' for sub in os.listdir(folder) if sub != '.DS_Store']

    for j in range(0, len(subfolders)):
        subfolder = subfolders[j]
        files = [f'{subfolder}/{file}' for file in os.listdir(subfolder) if file != '.DS_Store']

        data = [extract_data(file) for file in files]

        data_dict = {
            'records': data
        }

        with open(f"../data/img-extracts/{folder.split('/')[-2]}-{subfolder.split('/')[-1]}.json", 'w') as out_file:
            json.dump(data_dict, out_file)

        print(f'\t Completed subfolder {j}/{len(subfolders)}')

    print(f'Completed folder {i}/{len(folders)}')
```
