```python
import pandas as pd
import os
import io
import json
```

```python
def list_files_recursive(path):
    results = []
    for entry in os.listdir(path):
        full_path = os.path.join(path, entry)
        if os.path.isdir(full_path):
            results.extend(list_files_recursive(full_path))
        else:
            results.append(full_path)
    return results
```

```python
def make_dict_entry(file):
    if 'DS_Store' not in file:
        with open(file, 'r') as in_file:
            raw_text = in_file.read()

        return {
            'file': file,
            'url': file.replace('..', 'https://raw.githubusercontent.com/declanrjb/epstein-search/refs/heads/main/'),
            'text': raw_text,
            'sample': raw_text[0:200],
            'file_stem': file.split('/')[-1]
        }
```

```python
files = list_files_recursive('../data/text')
```

```python
records = [make_dict_entry(file) for file in files if '.DS_Store' not in file]
```

```python
df = pd.DataFrame(records)
```

```python
df.to_csv('../HouseFiles_TextAttachments_All.csv', index=False)
```

```python
with open('../HouseFiles_TextAttachments_All.json', 'w') as out_file:
    json.dump(records, out_file)
```
